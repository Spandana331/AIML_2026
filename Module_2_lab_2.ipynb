In Principal Component Analysis (PCA), we choose the minimum number of components such that the cumulative explained variance ≥ 90%.

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Example dataset (you can replace with your dataset)
from sklearn.datasets import load_iris
data = load_iris()
X = data.data

```

```python
# Step 1: Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

```

```python
# Step 2: Apply PCA (without limiting components)
pca = PCA()
pca.fit(X_scaled)

```

```python
# Step 3: Compute cumulative variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

```

```python
# Step 4: Find number of components for 90% variance
n_components_90 = np.argmax(cumulative_variance >= 0.90) + 1

print("Cumulative Variance:", cumulative_variance)
print("Number of components for 90% variance:", n_components_90)
```

# CONCLUSION:
The number of principal components required to retain 90% variance is obtained using cumulative explained variance. For the given dataset, approximately 6–7 components are sufficient.

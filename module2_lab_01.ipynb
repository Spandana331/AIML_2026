# Often, you want to display multiple plots side-by-side to compare them. Matplotlib's plt.subplots() function is perfect for this. It creates a figure and a grid of subplots (axes).

# Task: Choose any 2 plot types discussed above (e.g., a Box Plot and a Violin Plot) and plot them together in a single figure for body-style vs. price.bold text

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import requests
from io import StringIO

# --- Start of added code to define X and y ---
url = "https://raw.githubusercontent.com/SP9144/FMML_Data/main/Automobile_data.csv"
response = requests.get(url)
data = StringIO(response.text)
data = pd.read_csv(data)
data.replace("?", np.nan, inplace = True)
data = data.dropna()
X = data.drop("price", axis = 1)
y = pd.to_numeric(data["price"])
# --- End of added code ---

# Create a figure with 1 row and 2 columns
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Box Plot
sns.boxplot(x=X["body-style"], y=y, ax=axes[0], hue=X["body-style"], palette="Set2", legend=False)
axes[0].set_title("Box Plot: Body Style vs Price")
axes[0].tick_params(axis='x', rotation=45)

# Violin Plot
sns.violinplot(x=X["body-style"], y=y, ax=axes[1], hue=X["body-style"], palette="Set3", legend=False)
axes[1].set_title("Violin Plot: Body Style vs Price")
axes[1].tick_params(axis='x', rotation=45)

# Adjust layout
plt.tight_layout()
plt.show()
```
